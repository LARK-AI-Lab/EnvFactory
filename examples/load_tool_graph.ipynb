## Load and Save Tool Graph
Firstly, please configure your embedding and chat model in `.env`. For example:
```Python
EMBEDDING_URL=https://api.siliconflow.cn/v1/embeddings
EMBEDDING_API_KEY=...
EMBEDDING_MODEL=BAAI/bge-m3

CHAT_URL=https://api.deepseek.com
CHAT_API_KEY=...
CHAT_MODEL=deepseek-chat
```

````python
import os 
import json

# Load metadata
metadatas = []
metadata_path = "../envs/metadata"
metadata_files = []

with open("../configs/mcp_server.json") as f:
    servers = json.load(f)['mcpServers']
servers = [server for server in servers.keys()]

for server in servers:
    assert server + "_metadata.json" in os.listdir(metadata_path), f"{server}.json is not in metadata"
    metadata_files.append(server + "_metadata.json")

for file in metadata_files:
    with open(os.path.join(metadata_path, file), "r") as f:
        metadata = json.load(f)
    metadatas.append(metadata)
````

````python
from src.graph.tool_graph import ToolGraph

# Create tool graph
tool_graph = ToolGraph()
tool_graph.build_tool_graph(
    metadata = metadatas,
    enable_merge = False,
    enable_build_edge_with_llm = True,
)
````

````python
# Save to .pkl file
tool_graph.save("graph.pkl")
````
